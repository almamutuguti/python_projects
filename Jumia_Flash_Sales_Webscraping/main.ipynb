```python
import requests
import csv
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

service = Service()

driver = webdriver.Chrome(service=service)

csv_file = "sales.csv"
base_url = "https://www.jumia.co.ke/flash-sales/?page={}"

with open(csv_file, mode="w", encoding="utf-8", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Cover URL", "Name", "Price", "Old Price", "Rating", "Items Left"])

    page = 1
    
    while True:
    
        url = base_url.format(page)
        driver.get(url)
        time.sleep(3) #wait for page to load
        sales = driver.find_elements(By.CLASS_NAME, "prd")

        if not sales:
            print(f"All pages scarped")
            break

        for sale in sales:

            cover_url = sale.find_element(By.CLASS_NAME, "img").get_attribute("src")
            name = sale.find_element(By.CLASS_NAME, "name").text
            price = sale.find_element(By.CLASS_NAME, "prc").text
            if not price:
                print("Unable to retrieve the price")

            old_price = sale.find_element(By.CLASS_NAME, "prc").get_attribute("data-oprc")
            if not old_price:
                print("Unable to retrieve the old price")

            # rating = sale.find_element(By.CLASS_NAME, "stars").text
            # if not rating:
            #     print("Unable to retrieve the ratings")
            # rating = rating_tag["class"][1] if rating_tag and len(rating_tag["class"]) > 1 else "N/A"

            # items_left = sale.find_element(By.CLASS_NAME, "stk").text.strip()
            # if not items_left:
            #     print("Unable to retrieve the items left")

            # writer.writerow([cover_url, name, price, old_price, rating, items_left])
            writer.writerow([cover_url, name, price])

        page +=1
        time.sleep(2)

driver.quit()

print(f"All books saved to {csv_file}")
```
